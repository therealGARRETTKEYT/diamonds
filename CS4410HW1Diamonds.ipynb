```python
import pandas as pd
url = 'https://raw.githubusercontent.com/therealGARRETTKEYT/diamonds/main/diamonds.csv'
df = pd.read_csv(url)
```

```python
print(df.head(7))
```

```python
print(df.tail(7))
```

```python
print(df.describe())
```

```python
columns_to_select = ['cut' , 'color' , 'clarity']
selected_data = df[columns_to_select]
print(selected_data.describe())
```

```python
all_columns = pd.Series(['carat' , 'cut' , 'color' , 'clarity' , 'depth' , 'table' , 'price' , 'x' , 'y' , 'z'])
for column in all_columns:
  unique_values = df[column].unique()
  print(f"The unique values in column '{column}'are : {unique_values}")
```

```python
import matplotlib.pyplot as plt

numerical_columns = ['carat' , 'depth' , 'table' , 'price' , 'x' , 'y' , 'z']

df[numerical_columns].hist(bins=50, figsize=(10,10))
plt.tight_layout()
plt.show()
```
